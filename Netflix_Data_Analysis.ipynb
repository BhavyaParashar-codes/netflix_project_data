```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Plot settings
sns.set(style='whitegrid')
plt.rcParams['figure.figsize'] = (10,6)

# Load dataset
url = 'https://raw.githubusercontent.com/BhavyaParashar-codes/netflix_project_data/refs/heads/main/netflix_titles.csv'
df = pd.read_csv(url)

# Show data
#df.head()
#df.shape
#df.info()
(df.isnull().sum()/len(df))*100
df['director'] = df['director'].fillna('Unknown')
df['cast'] = df['cast'].fillna('Unknown')
df['country'] = df['country'].fillna('Unknown')
df['date_added'] = df['date_added'].fillna(df['date_added'].mode()[0])
df['rating'] = df['rating'].fillna(df['rating'].mode()[0])
df['duration'] = df['duration'].fillna(df['duration'].mode()[0])

sns.countplot(x='type', data=df,palette='pastel')
plt.title('Count of Movies and Yv Shows on Netflix')
plt.xlabel('Type')
plt.ylabel('Count')
plt.show()

top_country=df['country'].value_counts().head(10)
top_country.plot(kind='barh',color='red')
plt.title('Netflix Movies')
plt.xlabel('Name of titles')
plt.ylabel('country')
plt.gca().invert_yaxis()
plt.show()


df['date_added'] = df['date_added'].str.strip()
df['date_added']=pd.to_datetime(df['date_added'])
df['year_added']=df['date_added'].dt.year
df['year_added'].value_counts().sort_index().plot(kind='bar',figsize=(10,5),color='green')
plt.title('Netflix Content added Over the years')
plt.xlabel('year')
plt.ylabel('number of titles')
plt.show()

from collections import Counter

all_genres = ','.join(df['listed_in']).split(',')
top_genres = Counter([g.strip() for g in all_genres]).most_common(10)

genres_df = pd.DataFrame(top_genres, columns=['Genre', 'Count'])

sns.barplot(data=genres_df, x='Count', y='Genre', palette='magma')
plt.title('Top 10 Most Common Netflix Genres')
plt.show()
```

In this project, I explored a Netflix dataset using Python and performed the following steps:

Cleaned the data by handling missing values and formatting issues in date and categorical columns

Analyzed the distribution of content types (Movies vs TV Shows)

Identified top contributing countries

Explored how Netflix content evolved over the years

Extracted and visualized the most common genres

Key Insights:

Movies make up the majority of content on Netflix

The United States and India are the top content producers

Content additions started around 2014 and peaked  in 2019–2020

Internantional Movies,Dramas, Comedies, and Documentaries are the most dominant genres
